```python
# 1- Bir listeyi düzleştiren (flatten) fonksiyon yazın. Elemanları birden çok katmanlı listelerden, 
# ([[3],2] gibi) oluşabileceği gibi, non-scalar verilerden de oluşabilir. 
```

```python
#İlk Yöntem : 
def f(liste):
    if type(liste) is list:
        for i in liste:
            yield from f(i)
    else:
        yield liste

def flatten(liste):
    return list(f(liste))

liste = [[1,'a',['elif'],2],[[[3]],'özberk'],4,5]
flat_list = flatten(liste)
print(flat_list)
```

```python
# İkinci Yöntem : 
ilk_list = [1,3,5,[7,10],[10,12],["berk","öz"]]
bos_list = []


for i in ilk_list:
    if hasattr(i,"__iter__"):
        bos_list.extend(i)
    else:
        bos_list.append(i)
print(bos_list)

        
```
